```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data from excel files
df = pd.read_csv('KPA STATS 2011-2022.csv')

#remove empty rows
df = df.dropna(how='all')

#save file
df.to_csv('KPA STATS 2011-2022.csv', index=False)

print(df)
```

```python
# Remove empty columns
df = df.dropna(axis=1, how='all')

#save file
df.to_csv('KPA STATS 2011-2022.csv', index=False)

print(df)
```

```python
#Extracting data points(x,y coordinates from years and totals in each year)
columns = df.columns
Year_columns = []
for column in columns:
    Year_columns.append(column)

Year_columns.remove('STATUS ')
print(Year_columns)

totals = df.iloc[14]
totals_row = []
for total in totals:
    totals_row.append(total)

totals_row.remove('TOTAL ')
print(totals_row)

#print(len(Year_columns))
#print(len(totals_row))
```

## Plotting Container Traffic (TEU)


```python
## Setting x,y coordinates
Year_list = [int(x) for x in Year_columns]
Total_list = [int(x.replace(',', '')) for x in totals_row]
x = Year_list
y = Total_list

plt.figure(figsize=(5,3), dpi=300)

plt.plot(x,y)

ticksyaxis = [0,200000,400000,600000,800000,1000000,1200000,1400000,1600000]
plt.yticks(ticksyaxis, [f'{tick:,}' for tick in ticksyaxis])

years = [2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
plt.xticks(years, rotation=90, ha='right')

plt.title('Container Traffic (TEU)', fontdict={'fontname':'Comic Sans Ms', 'fontsize':20})
plt.xlabel('Years')
plt.ylabel('Container Traffic(TEU)')

# Display gridlines that separate each cell distinctly
plt.grid(which='major', color='grey', linestyle='-', linewidth=0.5)

plt.show()

```

```python
## Loading imports data into a list (imports_list)
importsvalues = df.iloc[2]
imports_row = []
for i in importsvalues:
    imports_row.append(i)
    
imports_row.remove('IMPORTS ')
imports_list = [int(x.replace(',', '')) for x in imports_row]
print(imports_list)

## Loading exports data into a list (exports_list)
exportsvalues = df.iloc[5]
exports_row = []
for e in exportsvalues:
    exports_row.append(e)
    
exports_row.remove('EXPORTS ')
exports_list = [int(x.replace(',', '')) for x in exports_row]
print(exports_list)

## Loading transhipment data into a list (trans_list)
transvalues = df.iloc[8]
trans_row = []
for t in transvalues:
    trans_row.append(t)
    
trans_row.remove('TRANSHIPMENT ')
trans_list = [int(x.replace(',', '')) for x in trans_row]
print(trans_list)
```

## Plotting Container Volume at Mombasa Port

```python
labels = Year_columns
import_values = imports_list
export_values = exports_list
transhipment_values = trans_list

plt.yticks(ticksyaxis, [f'{tick:,}' for tick in ticksyaxis])
plt.xticks(years, rotation=90, ha='right')

plt.bar(years, import_values, label='Imports')
plt.bar(years, export_values, bottom=import_values, label='Exports')
plt.bar(years, transhipment_values, bottom=np.array(export_values)+np.array(import_values), label='Transhipment')

plt.xlabel('Years')
plt.ylabel('Container Throughput(TEU)')
plt.title('Container Volume at Mombasa Port', fontdict={'fontname':'Comic Sans Ms', 'fontsize':20})

# Display gridlines that separate each cell distinctly
plt.grid(which='major', color='grey', linestyle='-', linewidth=0.5)
plt.legend()

plt.show()
```

## Plotting Container Traffic(TEU) of Imports, Exports and Transhipment

```python
## Setting coordinates for imports, exports and transhipment

plt.figure(figsize=(5,3), dpi=300)

plt.plot(x,imports_list,'red',label='Imports')
plt.plot(x,exports_list,'green',label='Exports')
plt.plot(x,trans_list,'blue',label='Transhipment')

plt.xticks(years, rotation=90)

plt.title('Imports, Exports and Transhipment', fontdict={'fontname':'Comic Sans Ms', 'fontsize':20})
plt.xlabel('Years')
plt.ylabel('Container Traffic(TEU)')

# Display gridlines that separate each cell distinctly
plt.grid(which='major', color='grey', linestyle='-', linewidth=0.5)

plt.legend()

plt.show()
```
